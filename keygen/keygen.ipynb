# Keygen

## Question
You are given a task to develop a simple keygen which require a two dimensional matrix with a size of N x N as an input, where N is an odd number between 3 and 9.
The matrix will be filled with integer numbers ranging between 0 and 49.
The following figure illustrates the input matrix with the size of 3 x 3.

![matrix](./img/matrix.png)

The expected output will be a key consist of 6 digits.
The first 3 digits could be generated by calculating the difference between the sum of its diagonals; or in other words the difference between the sum of D1 and the sum of D2.
The last 3 digits could be generated by calculating the difference between the sum of its crosses; or in other words the difference between the sum of C1 and the sum of C2.
